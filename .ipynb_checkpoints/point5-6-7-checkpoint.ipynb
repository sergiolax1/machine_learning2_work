```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from tensorflow.keras.datasets import mnist
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import train_test_split
from unsupervised import PCA_U ,TSNE_U 
```

#### 5 Train a naive logistic regression on raw MNIST images to distinguish between 0s and 8s. We are calling
#### this our baseline. What can you tell about the baseline performance?

```python
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# normalize images
train_images = train_images.astype(np.float32) / 255.0
test_images = test_images.astype(np.float32) / 255.0

# Merge training and test images and labels
X = np.concatenate((train_images, test_images), axis=0)
y = np.concatenate((train_labels, test_labels), axis=0)

# Select some random images to display
num_images = 50
indices = np.random.choice(range(X.shape[0]), num_images)
random_imgs = X[indices]
random_labels = y[indices]

# Visualizar las imgenes seleccionadas en una cuadrcula
fig = plt.figure(figsize=(10, 6))
for i in range(num_images):
    ax = fig.add_subplot(5, 10, i + 1)
    plt.imshow(random_imgs[i], interpolation="nearest", cmap=plt.cm.Greys_r)
    ax.set_title(int(random_labels[i]))
    ax.set_xticklabels([])
    ax.set_yticklabels([])

plt.tight_layout()
plt.show()

```

```python
train_images = train_images.reshape(-1, 28*28)
test_images = test_images.reshape(-1, 28*28)

#the images of zeros and eights are filtered
train_filter = np.where((train_labels == 0) | (train_labels == 8))
test_filter = np.where((test_labels == 0) | (test_labels == 8))

train_images, train_labels = train_images[train_filter], train_labels[train_filter]
test_images, test_labels = test_images[test_filter], test_labels[test_filter]

## train the data
X_train, X_test, y_train, y_test = train_test_split(train_images, train_labels, test_size=0.3, random_state=256)

### run and train the model
clf = LogisticRegression(random_state=345)
clf.fit(X_train, y_train)

### 
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

 accuracy and f1-scor are 99, that is, the regression model almost perfectly predicts the images

```python
num_images = 20
indices = np.random.choice(range(X_test.shape[0]), num_images)
random_imgs = X_test[indices]
true_labels = y_test[indices]
pred_labels = y_pred[indices]

fig = plt.figure(figsize=(20, 6))
for i in range(num_images):
    ax = fig.add_subplot(2, 10, i + 1)
    plt.imshow(random_imgs[i].reshape(28, 28), interpolation="nearest", cmap=plt.cm.Greys_r)
    ax.set_title(f"True: {true_labels[i]} | Pred: {pred_labels[i]}")
    ax.set_xticklabels([])
    ax.set_yticklabels([])

plt.tight_layout()
plt.show()
```

### unsupervised PCA

we apply the pca of the unsupervised library.
training data is reduced into two components with the intention of retraining and evaluating the logistic model with these two components.

```python
unsupervised_pca = PCA_U(n_components=2)
unsupervised_pca.fit(X_train)
X_train_unsupervised_pca = unsupervised_pca.transform(X_train)
X_test_unsupervised_pca = unsupervised_pca.transform(X_test)
```

```python
plt.scatter(X_train_unsupervised_pca[:, 0], X_train_unsupervised_pca[:, 1], c=y_train, cmap='viridis', s=5)
plt.xlabel('first component')
plt.ylabel('Second_component')
plt.colorbar()
plt.show()
```

it is observed that the two components are adequately separated, although a percentage is not adequately discriminated

```python
X_train_unsupervised_pca, X_test_unsupervised_pca, y_train_unsupervised_pca, y_test_unsupervised_pca =train_test_split(X_train_unsupervised_pca, 
                                                                                                     y_train,
                                                                                                 test_size=0.3, 
                                                                                                 random_state=128)
```

```python
cfl_with_unsupervised_pca = LogisticRegression(random_state=42)
cfl_with_unsupervised_pca.fit(X_train_unsupervised_pca, y_train_unsupervised_pca)

y_pred_unsupervised_pca = cfl_with_unsupervised_pca.predict(X_test_unsupervised_pca)
print(classification_report(y_test_unsupervised_pca, y_pred_unsupervised_pca ))
```

By training and evaluating the model in this way, the model prediction drops by 5%.

### unsupervised TSNE

For unsupervised tSNE,  we  choose a random sample. N=500 because tthis algoritm need too many procees time

```python
n_samples = 500
random_indices = np.random.choice(X_train.shape[0], n_samples, replace=False)
ramdom_indices_test= np.random.choice(X_test.shape[0], n_samples, replace=False)
```

```python
X_sample_train = X_train[random_indices]
y_sample_train = y_train[random_indices]
```

```python
unsupervised_tsne = TSNE_U()
X_train_unsupervised_tsne = unsupervised_tsne.fit_transform(X_sample_train)
```

```python
plt.scatter(X_train_unsupervised_tsne[:, 0], X_train_unsupervised_tsne[:, 1], c=y_sample_train, cmap='viridis', s=5)
plt.xlabel('first component')
plt.ylabel('Second_component')
plt.colorbar()
plt.show()
```

Now, let's train the logistic regression

```python
X_train_sample_tsne, X_test_sample_tsne, y_train_sample_tsne, y_test_sample_tsne = train_test_split(
    X_train_unsupervised_tsne, y_sample_train, test_size=0.2, random_state=42)


logistic_regression_tsne_unsupervised = LogisticRegression()
logistic_regression_tsne_unsupervised.fit(X_train_sample_tsne, y_train_sample_tsne)

y_test_pred_tsne_sample= logistic_regression_tsne_unsupervised.predict(X_test_sample_tsne)
# Calcular la precisin en el conjunto de validacin
print(classification_report(y_test_sample_tsne , y_test_pred_tsne_sample))
```

the separation of the components is very good , 
discriminates numbers very well
